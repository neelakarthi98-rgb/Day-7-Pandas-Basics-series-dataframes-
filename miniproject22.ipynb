###MOVIE DATA CLEANING###

```python
import pandas as pd
import numpy as np
data={'movie':['movie1','movie2','movie3','movie4'],
      'rating':[4.6,np.nan,4.4,np.nan],
       'release_year':['1994','2022',np.nan,'2006']} 
df=pd.DataFrame(data)
df.info()
```

###find the missing values 

```python
miss_value=df.isnull().sum()
print(miss_value)
```

//Handle missing Data//

```python
df['rating']=df['rating'].fillna(df['rating'].median())
print(df)
print()
df=df.dropna(subset=['release_year'])
print(df)
```

//correct Release year column into an integer//

```python
df.loc[:,'release_year'] = df['release_year'].astype(int)
print()
df.info()
```
