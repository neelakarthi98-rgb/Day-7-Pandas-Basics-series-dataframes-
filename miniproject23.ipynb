```python
import pandas as pd
import numpy as np
data={'student name':['alice','alice','bob','bob','charlie','charlie','alice','bob','charlie'],
      'subject':['tamil','english','tamil','english','tamil','english','maths','maths','maths'],
      'grades':[80,79,66,71,89,90,84,99,79]}
df=pd.DataFrame(data)
```

        ///calculate the average score of each subject////

```python
avg_subject=df.groupby('subject')['grades'].mean()
print("Average calcukation of each subject")
print(avg_subject)
```

////calculate the average score of each student///

```python
avg_student=df.groupby('student name')['grades'].mean()
print("Average calculation of each student")
print(avg_student)
```
