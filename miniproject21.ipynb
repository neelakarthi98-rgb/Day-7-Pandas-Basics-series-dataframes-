#SALES DATA ANALYSIS

objective:load a sample sales dataset and perform basic analysis to find total sales per product

```python
import pandas as pd
import numpy as np
data={'product':['A','B','A','C','B'],'quantity':[10,5,12,8,7],'price':[20,15,20,25,15]}
sales_df=pd.DataFrame(data)
print("original datas are:")
print(sales_df)

```

##//create a new column using vectorized operations

```python
sales_df['total_sales'] = sales_df['quantity'] * sales_df['price']


print(f"total sales values is{sales_df.total_sales}")
print()
print("updated value is:")
print(sales_df)
```

#calculate the product sales in each product

```python
product_sales=sales_df.groupby('product')['total_sales'].sum()
print("product sales is:")
print()
print(product_sales)
```
