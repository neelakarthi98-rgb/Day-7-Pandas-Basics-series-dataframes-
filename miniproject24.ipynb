////Customer purchase history///

```python
import pandas as pd
import numpy as np
data={'CustomerID': [101, 102, 103, 101, 104, 102, 101],
    'Product': ['Apple', 'Banana', 'Apple', 'Banana', 'Carrot', 'Apple', 'Carrot'],
    'Amount': [5.0, 2.5, 3.0, 2.5, 4.0, 3.0, 4.0]}
df=pd.DataFrame(data)
```

 ////find the total spending amount for the most popular product////

```python
Total_spen=df.groupby('CustomerID')['Amount'].sum()
print("a customer spending total amount is")
print(Total_spen)
print()
pop=df['Product'].value_counts()
print("popular product of each product")
print(pop)
most_pop=pop.max()
most_pop_pro=pop.idxmax()
print("most popular product")
print(most_pop_pro,most_pop)
```
